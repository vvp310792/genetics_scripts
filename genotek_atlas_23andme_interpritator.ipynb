```r
#определение LD

library(httr)
library(jsonlite)
library(xml2)
library(dplyr)
library(stringr)
library(googlesheets4)
mygenetics_genes<-read_sheet("https://docs.google.com/spreadsheets/d/1SusCweQyufxlX84VS4DWIWsxDnUM0LxH2dbwmpDoTho/edit#gid=1723433368",sheet ="Genes")


all_snp_data<-NULL
for (i in 1:length(mygenetics_genes$data)){


server <- "https://rest.ensembl.org"
ext <- paste0("/ld/human/",mygenetics_genes$data[i],"/1000GENOMES:phase_3:CEU?")
 
r <- GET(paste(server, ext, sep = ""), content_type("application/json"))
 

if (str_detect(content(r)[[1]],"Could not fetch variation object for id:")==FALSE) 
{# use this if you get a simple nested list back, otherwise inspect its structure
# head(data.frame(t(sapply(content(r),c))))
snp_data<-fromJSON(toJSON(content(r)))
snp_data$r2<-as.numeric(snp_data$r2)
snp_data<-snp_data %>%
filter(r2>0.32)%>%
select(variation1,variation2,r2,d_prime,population_name)%>%
arrange(desc(r2))


snp_data_string<-snp_data[1,]
snp_data_string$variation2<-snp_data_string$variation1
snp_data_string$r2<-1
snp_data_string$d_prime<-"1.000000"
snp_data<-rbind(snp_data_string,snp_data)
snp_data
all_snp_data<-rbind(all_snp_data,snp_data)
} else {
  print(mygenetics_genes$data[i])
}
} 
```

# Atlas report generate



```r
library(dplyr)
library(pegas)
library(googlesheets4)
library(data.table)
library(googledrive)
library(stringr)
options(stringsAsFactors = F)
options(datatable.fread.datatable=FALSE)
```

```r
mygenetics_polymorphism<-read_sheet("https://docs.google.com/spreadsheets/d/1SusCweQyufxlX84VS4DWIWsxDnUM0LxH2dbwmpDoTho/edit#gid=323829477",sheet = "LD_data_v2" )
rownames(mygenetics_polymorphism)<-mygenetics_polymorphism$variation2
mygenetics_polymorphism<-cbind(mygenetics_polymorphism,snp=rep(FALSE,nrow(mygenetics_polymorphism)),genotype=rep(FALSE,nrow(mygenetics_polymorphism)))
```

```r
data_example<-t(read.vcf("G:\\Общие диски\\R_and_D\\Проекты\\Интерпретация чужих данных\\Данные клиентов\\Виктория Борзова\\CG6148_vcf.vcf",from = 1, to = 10000000, which.loci = NULL, quiet = FALSE))
```

```r
data_example<-data_example %>%
as.data.frame() %>%
mutate(rsid=rownames(.))%>%
mutate(genotype=str_replace_all(.$cg6148,"/",""))%>%
select(rsid,genotype)
rownames(data_example)<-data_example$rsid
```

```r
data_example
```

```r
#Для atlas
data_example<-fread("G:\\Общие диски\\R_and_D\\Проекты\\Интерпретация чужих данных\\Данные клиентов\\Юлия Шабалова\\6987-4697.txt")
colnames(data_example)[1]<-"rsid"
rownames(data_example)<-atlas_row_data_example$rsid
```

```r
com_snp<-rownames(mygenetics_polymorphism)[rownames(mygenetics_polymorphism) %in% rownames(data_example)]
```

```r
mygenetics_polymorphism[com_snp,]$snp<-data_example[com_snp,]$rsid
mygenetics_polymorphism[com_snp,]$genotype<-data_example[com_snp,]$genotype
```

```r
mygenetics_polymorphism<-cbind(mygenetics_polymorphism,allele1_comp=mygenetics_polymorphism$"Allele 1"==mygenetics_polymorphism$"genotype",heterozygote_comp=mygenetics_polymorphism$Heterozygote==mygenetics_polymorphism$"genotype",allele2_comp=mygenetics_polymorphism$"Allele 2"==mygenetics_polymorphism$"genotype")
```

```r
mygenetics_polymorphism<-mygenetics_polymorphism %>%
mutate(all_allele=allele1_comp+allele2_comp+heterozygote_comp)
```

```r
mygenetics_polymorphism<-cbind(mygenetics_polymorphism,any(mygenetics_polymorphism$allele1_comp==TRUE,mygenetics_polymorphism$heterozygote_comp==TRUE,mygenetics_polymorphism$allele2_comp==TRUE)
```

```r
mygenetics_polymorphism <- as.data.frame(apply(mygenetics_polymorphism,2,as.character))
write.csv(mygenetics_polymorphism,"G:\\Общие диски\\R_and_D\\Проекты\\Интерпретация чужих данных\\Данные клиентов\\Виктория Борзова\\Виктория Борзова.csv")
```

```r
library(googledrive)
write.csv(mygenetics_polymorphism,"G:\\Общие диски\\R_and_D\\Проекты\\Интерпретация чужих данных\\Данные клиентов\\Юлия Шабалова\\Юлия Шабалова.txt")

poylymorphism_tempfile <- tempfile(pattern = "Юлия Шабалова", fileext = ".csv")
#> [1] "/var/folders/yx/3p5dt4jj1019st0x90vhm9rr0000gn/T//RtmpLtvjR3/iris-14cda64a59238.csv"
write.csv(mygenetics_polymorphism, poylymorphism_tempfile, row.names = FALSE)
poylymorphism_ss <- drive_upload(poylymorphism_tempfile, type = "spreadsheet")
```
